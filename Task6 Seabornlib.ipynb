```python
# Example

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

x=[21,15,9,6,31,3,47,4]
y=[20,8,34,12,29,2,62,1]
df=pd.DataFrame({'x':x,'y':y})
sns.lineplot(x='x',y='y',data=df)
plt.show()
```

```python
#Example

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x=np.array([21,3,17,39,4,45,6,72])
y=np.array([27,6,34,8,9,12,14,60])
df=pd.DataFrame(x,y)
sns.heatmap(df)
plt.show()
```

```python
# Pair plot
```

```python
#Example

import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('iris')
graph=sns.FacetGrid(df)
graph.map(plt.scatter,'sepal_length','petal_length')
plt.show()
```

```python
# Styling plot
```

```python
#Example

import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('iris')
sns.set_style('dark')
plt.figure(figsize=(10,3))
sns.boxplot(x='species',y='petal_length',data=df)
sns.despine()
plt.grid()
plt.show()
```
